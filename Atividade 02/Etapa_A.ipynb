```python
from random import randint
from os import listdir
import pandas as pd
import numpy as np
import cv2
```

```python
classes = ['circle', 'square', 'star', 'triangle']
dic = {}

for classe in classes:
    file = listdir(f'data/A/{classe}')

    randomImages = []
    randomIdx = []
    for i in range(0, 10):
        idx = randint(0, len(file) - 1)
        
        while idx in randomIdx:
            idx = randint(0, len(file) - 1)

        randomIdx.append(idx)
        randomImages.append(file[idx])
    
    dic[classe] = randomImages
```

```python
data = []

for classe in dic.keys():
    images = dic[classe]

    template = cv2.imread(f'data/A/{classe}/{dic[classe][0]}', cv2.IMREAD_GRAYSCALE)
    huMomentsTemplate = np.array(cv2.HuMoments(cv2.moments(template)).flatten())
    
    data.append([classe, dic[classe][0], 'Template', huMomentsTemplate[0], huMomentsTemplate[1], huMomentsTemplate[2], huMomentsTemplate[3], huMomentsTemplate[4], huMomentsTemplate[5], huMomentsTemplate[6], 0])

    for image in images[len(images)-3:len(images)]:
        img = cv2.imread(f'data/A/{classe}/{image}', cv2.IMREAD_GRAYSCALE)
        huMoments = np.array(cv2.HuMoments(cv2.moments(img)).flatten())

        data.append([classe, image, 'Comparacao', huMoments[0], huMoments[1], huMoments[2], huMoments[3], huMoments[4], huMoments[5], huMoments[6], np.linalg.norm(huMomentsTemplate-huMoments)])

entreMesmasClasses = pd.DataFrame(data, columns=['Classe', 'Imagem', 'Tipo', 'Hu1', 'Hu2', 'Hu3', 'Hu4', 'Hu5', 'Hu6', 'Hu7', 'Distancia euclidiana'])

entreMesmasClasses
```

```python
for c in entreMesmasClasses['Classe'].unique().tolist():
    print(f'Classe: {c}')
    display(entreMesmasClasses[entreMesmasClasses['Classe'] == c])
```

```python
entreMesmasClasses['Distancia euclidiana'].sum(), np.std(entreMesmasClasses['Distancia euclidiana'])
```

```python
data = []

for idx, classe in enumerate(dic.keys()):
    template = dic[classe][0]

    template = cv2.imread(f'data/A/{classe}/{template}', cv2.IMREAD_GRAYSCALE)
    huMomentsTemplate = np.array(cv2.HuMoments(cv2.moments(template)).flatten())

    data.append([idx, classe, dic[classe][0], 'Template', huMomentsTemplate[0], huMomentsTemplate[1], huMomentsTemplate[2], huMomentsTemplate[3], huMomentsTemplate[4], huMomentsTemplate[5], huMomentsTemplate[6], 0])

    for classeComparacao in dic.keys():

        if classe != classeComparacao:
            imagesComparacao = dic[classeComparacao]

            for image in imagesComparacao[len(imagesComparacao)-3:len(imagesComparacao)]:
                img = cv2.imread(f'data/A/{classeComparacao}/{image}', cv2.IMREAD_GRAYSCALE)
                huMoments = np.array(cv2.HuMoments(cv2.moments(img)).flatten())

                data.append([idx, classeComparacao, image, 'Comparacao', huMoments[0], huMoments[1], huMoments[2], huMoments[3], huMoments[4], huMoments[5], huMoments[6], np.linalg.norm(huMomentsTemplate-huMoments)])

entreClassesDiferentes = pd.DataFrame(data, columns=['Iteracao', 'Classe', 'Imagem', 'Tipo', 'Hu1', 'Hu2', 'Hu3', 'Hu4', 'Hu5', 'Hu6', 'Hu7', 'Distancia euclidiana'])

entreClassesDiferentes
```

```python
for idx in entreClassesDiferentes['Iteracao'].unique().tolist():
    display(entreClassesDiferentes[entreClassesDiferentes['Iteracao'] == idx])
```
